```python
import gender_guesser.detector as gender
```

```python
import pandas as pd
```

```python
d = gender.Detector(case_sensitive=False)
```

```python
df_all_un = pd.read_csv('amazon_undecided.csv', names=['ReviewerID', 'ProductID', 'Name', 'Gender','Helpfulness', 'Review', 'Rating','Timestamp'],sep='|')
```

```python
df_all_un.head()
```

```python
import string
translator = str.maketrans('', '', string.punctuation)
```

```python
df_sample = df_all_un.sample(1000)
```

```python
def get_gender(name):
    try:
        n = name.lower()
        if ('mom' in n) or ('girl' in n) or ('angel' in n) or ('mum' in n) or ('mother' in n) or ('woman' in n):
            return 'female'
        if ('boy' in n) or ('dude' in n):
            return 'male'
        temp = name.translate(translator).split()
    except:
        return 'unknown'
    
    if len(temp)>0:
        first_name = temp[0]
        first_name = ''.join([i for i in first_name if not i.isdigit()])
        gender = d.get_gender(first_name,'usa')
        return gender
    else:
        return 'unknown'
    
```

```python
df_sample['Gender'] = df_sample['Name'].apply(lambda x:get_gender(x))
```

```python
df_sample.sample(100)
```
