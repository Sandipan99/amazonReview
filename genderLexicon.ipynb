```python
import pandas as pd
```

```python
dict_lexicon = {}
with open('emnlp14gender.csv') as fs:
    for line in fs:
        try:
            word,score = line.strip().split(',')
            dict_lexicon[word.replace("\"",'')] = float(score.replace("\"",''))
        except:
            pass
```

```python
dict_lexicon
```

```python
from nltk.tokenize import word_tokenize
import string
from collections import Counter
from tqdm import tqdm
```

```python
puncts = string.punctuation
```

```python
puncts+='``'
```

```python
with open('inference_lexicon_amazon','w') as ft:
    ft.write('True_label,Predicted_label,ReviewerID\n')
    with open('User_level_test_with_id.csv') as fs:
        for line in tqdm(fs,total=7664045,position=0):
            line = line.strip()
            ind = line.find(',')
            id_ = line[:ind]
            review = line[ind+1:-2]
            label = line[-1]
            words = word_tokenize(review)
            words = [w for w in words if w not in puncts]
            word_cnt = Counter()
            sent_len = 0
            for w in words:
                if w not in puncts:
                    word_cnt[w]+=1
                    sent_len+=1
            score = 0.0
            for w in word_cnt:
                if w in dict_lexicon:
                    score+=dict_lexicon[w]*(word_cnt[w]/sent_len)
            if score>0:
                ft.write(label+',1,'+id_)
            else:
                ft.write(label+',0,'+id_)
            ft.write('\n')    
            
```
